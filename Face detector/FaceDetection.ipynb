`````````````````````````````````````````````
import cv2
import torch
`````````````````````````````````````````````

`````````````````````````````````````````````
#loading the cascades
face_cascade =  cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye_cascade =  cv2.CascadeClassifier("haarcascade_eye.xml")
smile_cascade = cv2.CascadeClassifier("haarcascade_smile.xml")
`````````````````````````````````````````````

`````````````````````````````````````````````
#Defining a function that will do the detections
def detect(gray, frame):
  
  faces = face_cascade.detectMultiScale(gray,1.3, 5 ) #detecting boundaries of detection box
  for (x, y, w, h) in faces:
    cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0), 2) #draw face box #image, upper left corner, bottom right corner, box color, image thickness
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = frame[y:y+h, x:x+w]
    
    eyes = eye_cascade.detectMultiScale(roi_gray,1.1, 22) #detecting boundaries of detection box #picture, scaling factor, number of neighbors
    for (ex, ey, ew, eh) in eyes:
      cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (0,255,0), 2) #draw eye boxes #image, upper left corner, bottom right corner, box color, image thickness
    
    smile = smile_cascade.detectMultiScale(roi_gray,1.7, 22)
    for (sx, sy, sw, sh) in smile:
      cv2.rectangle(roi_color, (sx,sy), (sx+sw, sy+sh), (0,0,255),2)

  return frame
`````````````````````````````````````````````

`````````````````````````````````````````````
#apply detect fucntion on webcam
video_capture = cv2.VideoCapture(0) #0 for webcam from internal webcam, 1 for external webcam
while True:
  _, frame = video_capture.read()
  gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
  canvas = detect(gray, frame)
  cv2.imshow('Video', canvas)
  if cv2.waitKey(1) & 0xFF == ord('q'):````````````````````````````````````````````
    break

video_capture.release()
cv2.destroyALLWindows()
`````````````````````````````````````````````
